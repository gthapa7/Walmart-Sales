# Model Comparison: Random Forest, XGBoost, and LightGBM
This notebook compares the performance of three models trained for Walmart Weekly Sales forecasting:
- Random Forest
- XGBoost
- LightGBM

For each model, we will:
- Load the saved model
- Load the test data and features
- Generate predictions
- Calculate evaluation metrics (RMSE, MAE, RÂ²)
- Summarize results in a comparison table and plot

```python
# --- Manual Model Outcome Comparison ---
import pandas as pd
import matplotlib.pyplot as plt

# Enter your outcome metrics for each model below (updated with actual values)
results = pd.DataFrame({
    'Model': ['Random Forest', 'XGBoost', 'LightGBM'],
    'RMSE': [77731.47, 77868.17, 50887.93],
    'MAE': [55039.90, 51860.75, 37220.34],
    'R2': [0.9790, 0.9789, 0.9908]
})

print(results)

# Bar plot for RMSE comparison
plt.figure(figsize=(7,4))
plt.bar(results['Model'], results['RMSE'], color=['royalblue','orange','green'])
plt.title('Model Comparison: RMSE')
plt.ylabel('RMSE')
plt.show()

# Bar plot for MAE comparison
plt.figure(figsize=(7,4))
plt.bar(results['Model'], results['MAE'], color=['royalblue','orange','green'])
plt.title('Model Comparison: MAE')
plt.ylabel('MAE')
plt.show()

# Bar plot for RÂ² comparison
plt.figure(figsize=(7,4))
plt.bar(results['Model'], results['R2'], color=['royalblue','orange','green'])
plt.title('Model Comparison: RÂ²')
plt.ylabel('RÂ²')
plt.ylim(0,1)
plt.show()
```

## About the Evaluation Metrics Used
The RMSE, MAE, and RÂ² values shown below were calculated from each model's predictions on the same test set. For each model (Random Forest, XGBoost, and LightGBM), predictions were generated using the trained model, and the following metrics were computed:

- **RMSE (Root Mean Squared Error):** Measures the average magnitude of the prediction error, penalizing larger errors more heavily.
- **MAE (Mean Absolute Error):** Measures the average absolute difference between predicted and actual values.
- **RÂ² (Coefficient of Determination):** Indicates the proportion of variance in the target variable explained by the model (closer to 1 is better).

These metrics provide a fair and direct comparison of model performance on unseen data.
